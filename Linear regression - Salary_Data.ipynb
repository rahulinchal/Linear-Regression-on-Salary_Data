```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("C:/Users\Rahul\Desktop\Data_Science\Youtube\csv/Salary_Data.csv")
x = data.iloc[:,:-1].values 
y = data.iloc[:,1].values
y
```

```python
sns.barplot(x = "YearsExperience", y = "Salary", data = data )
```

```python
correlation = data.corr()
sns.heatmap(correlation, annot = True)
```

# Split data into training and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 1/3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

# Fit Linear Regression into Training data set 

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

# Predict the test results

```python
y_pred = lr.predict(x_test)
y_pred
```

# Visualise the test set results 

```python
plt.scatter(x_test, y_test, color = "blue")
plt.plot(x_train, lr.predict(x_train), color = "Red")
plt.xlabel("Salary")
plt.ylabel("years")
plt.title("predicted comparision")
```

# Calculating the errors

```python
from sklearn import metrics
print("MAE : ", metrics.mean_absolute_error(y_test, y_pred))
print("MSE : ", metrics.mean_squared_error(y_test, y_pred))
print("RMSE : ", np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
```

```python
lr.score(x_train, y_train), lr.score(x_test, y_test)
```
